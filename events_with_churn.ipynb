```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/events_features_extended.csv')
df1 = df1.rename(columns={'Unnamed: 0': 'user_id'})
df1
```

```python
df1.columns
```

```python
df2 = pd.read_csv('data/user_churn_risk_analysis.csv')
df2.drop(columns=['Unnamed: 0'], inplace=True)
df2
```

```python
df2.columns
```

### Merging both

```python
merged_df = pd.merge(df1, df2[['user_id', 'churn_score', 'risk_category', 
                                'is_likely_churn', 'risk_velocity', 
                                'primary_risk_drivers', 'estimated_days_to_churn']], 
                                on='user_id', how='left')
```

```python
merged_df.columns
```

```python
merged_df
```

```python
merged_df.to_csv('data/events_with_churn_score.csv', index=False)
```
